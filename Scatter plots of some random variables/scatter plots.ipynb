```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create data
N = 500
x = np.random.rand(N)
y = np.random.rand(N)
colors = (0,0,0)
area = np.pi*3

# Plot of independent random variables
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.title('Scatter plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Create data

import random

x = []
for i in range(100):
    x_new = random.randint(500)
    x.append(x_new)
    y_new = 1000
    while (y_new > x_new).any():
        y_new = random.randint(500-x)
        
    y.append(y_new)
        
colors = (0,0,0)
area = np.pi*3

# Plot
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.title('Scatter plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Create data

from numpy import random

x = []
y = []
for i in range(1000):
    x_new = random.randint(500)
    x.append(x_new)
    y_new = 1000
    while (x_new-250)**2 + (y_new-250)**2 > 500**2:
        y_new = random.randint(500-x_new)
        
    y.append(y_new)
        
colors = (0,0,0)
area = np.pi*3

# Plot
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.title('Scatter plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
